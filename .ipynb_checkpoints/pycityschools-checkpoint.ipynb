```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
print(school_data_complete)
```

```python
#Calculate the total number of schools 
count_school_name = school_data_complete[school_data_complete.columns[4]].value_counts()
#print(len(count_school_name))

#Calculate the total number of students
count_total_student = school_data_complete[school_data_complete.columns[0]].count()
#print(count_total_student)

#Calculate the total budget
count_total_budget = (school_data_complete[school_data_complete.columns[10]].unique().sum())
#print(count_total_budget)

#Calculate the average maths score
count_average_math = school_data_complete[school_data_complete.columns[6]].mean()
#print(count_average_math)

#Calculate the average reading score
count_average_reading = school_data_complete[school_data_complete.columns[5]].mean()
#print(count_average_reading)

#Calculate the percentage of students with a passing maths score (50 or greater)
percent_math_pass = ((school_data_complete['maths_score'].where(school_data_complete['maths_score'] >= 50).count() / count_total_student)*100)
#print(percent_math_pass)

#Calculate the percentage of students with a passing reading score (50 or greater)
percent_reading_pass = ((school_data_complete['reading_score'].where(school_data_complete['reading_score'] >= 50).count() / count_total_student)*100)
#print(percent_reading_pass)

#Calculate the percentage of students who passed maths and reading (% Overall Passing)
percent_overall_pass = ((school_data_complete['student_name'].loc[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score']>=50)].count() / count_total_student)*100)
#print(percent_overall_pass)


#print(school_data_complete['reading_score'].where(school_data_complete['reading_score'] >= 50).count())
#print(school_data_complete['maths_score'].where(school_data_complete['maths_score'] >= 50).count())
```

```python
LGA_summary_df = pd.DataFrame({'Total Schools':[len(count_school_name)],
                            'Total Students':[count_total_student],
                            'Total Budget':[count_total_budget],
                            'Average Maths Score':[count_average_math],
                            'Average Reading Score':[count_average_reading],
                            '% Passing Maths':[percent_math_pass],
                            '% Passing Reading':[percent_reading_pass],
                            '% Overall Passing':[percent_overall_pass]})
LGA_summary_df.loc[:, "Total Students"] = LGA_summary_df["Total Students"].map('{:,}'.format)
LGA_summary_df.loc[:, "Total Budget"] = LGA_summary_df["Total Budget"].map('${:,.2f}'.format)
#print(LGA_summary_df.dtypes)
LGA_summary_df 
```

```python
school_group_by = school_data_complete.groupby(['school_name'])

school_names = school_data_complete['school_name'].sort_values().unique()
#print(school_names)

school_type =  school_data.sort_values(['school_name'])['type']
#print(school_type)

total_student_ps = school_group_by['student_name'].count()
#print(total_student_ps.dtypes)
total_student_ps_list = list(school_group_by['student_name'].count())

total_school_budget = (school_group_by['budget'].sum()/total_student_ps)
# total_school_budget1 = (school_group_by['budget'].unique())
#print(total_school_budget.dtypes)
total_school_budget_list = list(school_group_by['budget'].sum()/total_student_ps)

ps_budget = (total_school_budget/total_student_ps)
#print(ps_budget)
ps_budget_list = list((total_school_budget/total_student_ps))
# ps_budget_list = ps_budget_list.astype(str).astype(int)

avg_maths_ps = (school_group_by['maths_score'].mean())
avg_maths_ps_list = list(school_group_by['maths_score'].mean())
#print(avg_maths_ps)

avg_reading_ps = (school_group_by['reading_score'].mean())
avg_reading_ps_list = list(school_group_by['reading_score'].mean())
#print(avg_reading_ps)


maths_group_by = school_data_complete[school_data_complete['maths_score'] >= 50].groupby(['school_name'])
percent_math_pass_ps = ((maths_group_by['maths_score'].count() / total_student_ps)*100)
percent_math_pass_ps_list = list((maths_group_by['maths_score'].count() / total_student_ps)*100)
#print(percent_math_pass_ps)
readings_group_by = school_data_complete[school_data_complete['reading_score'] >= 50].groupby(['school_name'])
percent_reading_pass_ps = ((readings_group_by['reading_score'].count() / total_student_ps)*100)
percent_reading_pass_ps_list = list((readings_group_by['reading_score'].count() / total_student_ps)*100)
#print(percent_reading_pass_ps)
overall_group_by = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby(['school_name'])
percent_overall_pass_ps = ((overall_group_by['student_name'].count() / total_student_ps)*100)
percent_overall_pass_ps_list = list((overall_group_by['student_name'].count() / total_student_ps)*100)
#print(percent_overall_pass_ps)



School_summary_df = pd.DataFrame({'':school_names,
                                  'School Type':school_type,
                                  'Total Students':total_student_ps_list,
                                  'Total School Budget':total_school_budget_list,
                                  'Per Student Budget':ps_budget_list,
                                  'Average Math Score':avg_maths_ps_list,
                                  'Average Reading Score':avg_reading_ps_list,
                                  '% Passing Maths':percent_math_pass_ps_list,
                                  '% Passing Reading':percent_reading_pass_ps_list,
                                  '% Overall Passing':percent_overall_pass_ps_list}).set_index('')

School_summary_df.loc[:, 'Total School Budget'] = School_summary_df['Total School Budget'].map('${:,.2f}'.format)
School_summary_df.loc[:, 'Per Student Budget'] = School_summary_df['Per Student Budget'].map('${:,.2f}'.format)
#School_summary_df['Per Student Budget'] = School_summary_df.apply(lambda x: np.array(x['Per Student Budget']).astype(float),axis=1)

# print(School_summary_df.dtypes)
School_summary_df
```

```python
#create the bins in which data will be held
spending_bins = [0, 585, 630, 645, 680]
#create the names for the four bins
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending_df = pd.DataFrame({'':school_names,
                                  'School Type':school_type,
                                  'Total Students':total_student_ps_list,
                                  'Total School Budget':total_school_budget_list,
                                  'Per Student Budget':ps_budget_list,
                                  'Average Maths Score':avg_maths_ps_list,
                                  'Average Reading Score':avg_reading_ps_list,
                                  '% Passing Maths':percent_math_pass_ps_list,
                                  '% Passing Reading':percent_reading_pass_ps_list,
                                  '% Overall Passing':percent_overall_pass_ps_list}).set_index('')
print(school_spending_df.dtypes)

school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_spending_df['Per Student Budget'], spending_bins, labels=labels)

school_spending_df

spending_maths_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

spending_summary = pd.DataFrame({
          "Average Maths Score" : spending_maths_scores,
          "Average Reading Score": spending_reading_scores,
          "% Passing Maths": spending_passing_maths,
          "% Passing Reading": spending_passing_reading,
          "% Overall Passing": overall_passing_spending})

spending_summary

spending_summary["Average Maths Score"] = spending_summary["Average Maths Score"].map("{:.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:.2f}".format)
spending_summary["% Passing Maths"] = spending_summary["% Passing Maths"].map("{:.2f}".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].map("{:.2f}".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].map("{:.2f}".format)

spending_summary
```
